# Yearly Data 

```python
def yearly_data(year):
    year_info = df_copy[df_copy['Year'] == year]
    
    year_data_list = [
        year_info['Confirmed'].sum(),
        year_info['Confirmed'].mean(),
        year_info['Cured'].sum(),
        year_info['Cured'].mean(),
        year_info['Deaths'].sum(),
        year_info['Deaths'].mean(),
        year_info['Confirmed'].sum()-(year_info['Cured'].sum()+year_info['Deaths'].sum()),
        "{:.2f}%".format((year_info['Cured'].sum()/year_info['Confirmed'].sum())*100),
        "{:.2f}%".format((year_info['Deaths'].sum()/year_info['Confirmed'].sum())*100),
        "{:.2f}%".format(((year_info['Confirmed'].sum()-(year_info['Cured'].sum()+year_info['Deaths'].sum()))/year_info['Confirmed'].sum())*100)
    ]
    
    return year_data_list
```

```python
yearly_info = {
    'Stats':[
        'total_confirmed_cases',
        'average_cases',
        'total_cured',
        'average_cured',
        'total_no_of_deaths',
        'average_deaths',
        'total_active_cases',
        'recovery_rate',
        'death_rate',
        'active_cases_rate'
    ],
    '2020':yearly_data(year = 2020),
    '2021':yearly_data(year = 2021)
}
```

```python
yearly_info_df = pd.DataFrame(yearly_info)
```

```python
yearly_info_df
```

# Monthly data 

```python
def monthly_details(month,year):
    year_info = df_copy[df_copy['Year'] == year]
    month_info = year_info[year_info['Month'] == month]

    month_data_list = [
        month_info['Confirmed'].sum(),
        month_info['Confirmed'].mean(),
        month_info['Cured'].sum(),
        month_info['Cured'].mean(),
        month_info['Deaths'].sum(),
        month_info['Deaths'].mean(),
        month_info['Confirmed'].sum()-(month_info['Cured'].sum()+month_info['Deaths'].sum()),
        "{:.2f}%".format((month_info['Cured'].sum()/month_info['Confirmed'].sum())*100),
        "{:.2f}%".format((month_info['Deaths'].sum()/month_info['Confirmed'].sum())*100),
        "{:.2f}%".format(((month_info['Confirmed'].sum()-(month_info['Cured'].sum()+month_info['Deaths'].sum()))/month_info['Confirmed'].sum())*100)
    ]
    
    return month_data_list
```

### Monthly data of 2020 

```python
monthly_data_2020 = {
    'stats':[
        'Confirmed cases',
        'Average confirmed cases',
        'Cured',
        'Average cured',
        'Deaths',
        'Average deaths',
        'total active cases',
        'recovery rate',
        'death rate',
        'active cases rate',
    ],
}
```

```python
months = ['Jan','feb','March','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
for i in range(12):
    monthly_data_2020[months[i]] = monthly_details(month = i+1 , year = 2020)
```
